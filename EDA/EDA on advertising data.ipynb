```python
def outlier_removal(data,col):
    plt.figure(figsize=(20,20))
    fig,axes=plt.subplots(ncols=1,nrows=2,figsize=(12,4))
    sns.boxplot(data=data,x=col,ax=axes[0],color="red").set_title("Before Outlier removal"+col)
    lower_limit=data[col].mean()-(3*data[col].std())
    upper_limit=data[col].mean()+(3*data[col].std())
    a=[]
    for x in data[col]:
        if x<lower_limit:
            x=lower_limit
            a.append(x)
        elif x>upper_limit:
            x=upper_limit
            a.append(x)
        else:
            x=x
            a.append(x)
    data[col]=a
    sns.boxplot(data=data,x=col,ax=axes[1],color='blue').set_title("After Outlier Removal"+col)
    plt.show()
    return data
    
```

```python
for col in data.columns:
    data=outlier_removal(data,col)
plt.show()
```
